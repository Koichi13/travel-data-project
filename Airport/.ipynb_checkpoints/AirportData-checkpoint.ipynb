```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../flight-delays/flights.csv")
df["total_delay"] = df["DEPARTURE_DELAY"] + df["ARRIVAL_DELAY"]
df.head()
```

```python
airport_delay = df[["ORIGIN_AIRPORT", "DEPARTURE_DELAY"]]
airport_delay
```

```python
df_airport_clean = airport_delay[airport_delay['DEPARTURE_DELAY'] >= 0.0]
airport_count = df_airport_clean["ORIGIN_AIRPORT"].value_counts().nlargest(10)
```

```python
airport_count = pd.DataFrame(airport_count)
airport_count
```

```python
airport_count.plot(kind="bar", facecolor="skyblue", alpha=0.75, align="center", rot=45)
plt.title("Top 10 Most Delayed Airports")
plt.xlabel("Airport")
plt.ylabel("Number of Delays")
plt.tight_layout()
```

```python
airport_delay_2 = df[["DESTINATION_AIRPORT", "ARRIVAL_DELAY"]]
airport_delay_2 
```

```python
df_airport_clean_2 = airport_delay_2[airport_delay_2['ARRIVAL_DELAY'] >= 0.0]
airport_count_2 = df_airport_clean_2["DESTINATION_AIRPORT"].value_counts().nlargest(10)
```

```python
airport_count_2 = pd.DataFrame(airport_count_2)
airport_count_2
```

```python
airport_count_2.plot(kind="bar", facecolor="skyblue", alpha=0.75, align="center", rot=45)
plt.title("Top 10 Most Delayed Airports")
plt.xlabel("Airport")
plt.ylabel("Number of Delays")
plt.tight_layout()
```
